```python
# Create new Semi-minor axis length feature
df_ast_2['Semi-minor axis (AU)'] =  np.sqrt(df_ast_2['a'] * df_ast_2['Semilatus_rectum'])
df_ast_2[['a', 'Semi-minor axis (AU)', 'e']].head()
```

```python
def calc_orbit_circum(a, b):
    # Create a function that calculates approx. circumference of elliptical orbit from semi-major and minor axes lengths
    h = ((a - b)**2)/((a + b)**2)
    
    c = np.pi*(a + b)*(1 + 3*h/(10 + np.sqrt(4 - 3*h)))
    
    return c
    
```

```python
# Create Approx Orbit length feature using function created above
df_ast_2['Approx_Orbit_Length'] = df_ast_2.apply(lambda row: calc_orbit_circum(row['a'], row['Semi-minor axis (AU)']), axis = 1)
df_ast_2.head()
```

```python
df_ast_2[['a', 'Semi-minor axis (AU)', 'Approx_Orbit_Length']].head()
```

```python
# Create another new feature, Approx average orbital velocity
df_ast_2['Approx_Avg_Orb_Vel (AU/yr)'] = df_ast_2['Approx_Orbit_Length']/df_ast_2['Orbital_period']
df_ast_2.head()
```

```python
def calc_vel(r, a):
    # A function to calculate instantaneous velocity at any point in orbit
    G = 6.67e-11  # Gravitational constant
    M_sun = 1.989e+30    # Mass of sun
    AU = 1.5e+11   # One Astronomical unit in meters
    
    v = np.sqrt(G*M_sun*(2/(r*AU) - 1/(a*AU)))   # Calculate velocity in m/s
    
    sec_year = 3.154e+7    # Number of seconds in year
    meters_AU = 1.496e+11      # Number of meters in AU
                
    v_convert = (v*sec_year)/meters_AU    # Convert to AU/yr
    
    return v_convert
```

```python
# Create two new numeric features using function above, velocity at perihelion and aphelion (essentially max and min velocity)
df_ast_2['Velocity_peri'] = df_ast_2.apply(lambda row: calc_vel(row['Perihelion_dist'], row['Semi_major_axis']), axis=1)
df_ast_2['Velocity_aphi'] = df_ast_2.apply(lambda row: calc_vel(row['Aphelion_dist'], row['Semi_major_axis']), axis=1)

df_ast_2.head()
```
